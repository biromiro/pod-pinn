```python
result_files = {
    'supervised': 'accuracy_res_time.json',
    'unsupervised': 'accuracy_res_time_unsupervised.json',
    #'mixed_0.01': 'accuracy_res_time_mixed_0.01.json',
    'mixed_0.05': 'accuracy_res_time_mixed_0.05.json',
    #'mixed_0.1': 'accuracy_res_time_mixed_0.1.json',
    #'mixed_0.2': 'accuracy_res_time_mixed_0.2.json',
    #'mixed_0.3': 'accuracy_res_time_mixed_0.3.json',
    #'mixed_0.4': 'accuracy_res_time_mixed_0.4.json',
    'mixed_0.5': 'accuracy_res_time_mixed_0.5.json',
    #'mixed_0.75': 'accuracy_res_time_mixed_0.75.json',
    'mixed_1': 'accuracy_res_time_mixed_1.json',
    #'mixed_2': 'accuracy_res_time_mixed_2.json',
    #'mixed_5': 'accuracy_res_time_mixed_5.json',
    #'mixed_10': 'accuracy_res_time_mixed_10.json'
}
```

```python
accuracy = {}
residual = {}
time = {}

for key, value in result_files.items():
    with open(value, 'r') as f:
        data = f.read()
        data = eval(data)
        accuracy[key] = data['accuracy']
        residual[key] = data['residual']
        time[key] = data['time']
        
print(accuracy)
print(residual)
print(time)
```

```python
# Plot accuracy values using log scale
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.set_yscale('log')

# set y-axis domain to 0 to 0.1
for key, value in accuracy.items():
    ax.plot(value, label=key)
    
ax.set(xlabel='Epoch', ylabel='Accuracy',
         title='Accuracy')
ax.grid()
ax.legend()
plt.show()
```

```python
# Plot residual values using log scale
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.set_yscale('log')

# set y-axis domain to 0 to 0.1
for key, value in residual.items():
    ax.plot(value, label=key)
    
ax.set(xlabel='Epoch', ylabel='Residual',
         title='Residual')
ax.grid()
ax.legend()
plt.show()
```
